##Cálculo de Média

```python
nota1 = int(input("Informe a nota do bimestre 1 (0-100): "))
```

```python
nota2 = int(input("Informe a nota do bimestre 2 (0-100): "))
```

```python
media = (nota1 + nota2) / 2
```

```python
if media >= 50:
    print ("Aprovado")
else:
    print("Reprovado")
```
